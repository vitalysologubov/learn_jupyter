```python
import csv
```

```python
streets = {}
```

```python
%time
with open('bus_stops.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```

```python
%time
sorted(streets.items(), key=lambda item: item[1], reverse=True)
```
